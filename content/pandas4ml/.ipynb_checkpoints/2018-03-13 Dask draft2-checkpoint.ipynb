```python
import dask.bag as db
import dask.diagnostics as dg
import pandas as pd
import json
import numpy as np
from toolz.curried import map, pipe, get, compose, pluck, valmap
from bokeh.io import output_notebook

output_notebook()
```

```python
MEGABYTES=1024**2
```

```python
DATA_DIR = '/home/severo/.kaggle/competitions/how-good-is-your-medium-article/'
files = !ls {DATA_DIR}
datasets = {f.split('.')[0]:DATA_DIR + f for f in files}
datasets
```

```python
!cat {datasets['train']} | tail -n1 | jq 'del(.content)' | jq .
```

```python
with open(datasets['train'], 'r') as f:
    data = []
    for line in f:
        parsed = json.loads(line)
        data.append({
            'published_timestamp': parsed.get('published', {}).get('$date'),
            'read_time': parsed.get('meta_tags', {}).get('twitter:data1'),
            'url': parsed.get('_id')
        })

df = (
    pd
    .DataFrame(data)
    .assign(published_timestamp = lambda df: pd.to_datetime(df['published_timestamp']))
    .assign(read_time = lambda df: df['read_time'].str.split(' ').str[0].astype(int))
    .set_index('published_timestamp')
    .sort_index()
)

df.head()
```

```python
(
    df['2015':]
    .groupby(lambda index: pd.to_datetime(index.strftime("%Y/%m")))
    .aggregate([np.mean, np.std])
    .plot(figsize=(12,5), style='o--', subplots=True)
);
```

```python
with dg.Profiler() as prof, dg.ResourceProfiler() as rprof, dg.ProgressBar():
    dag = (
        db
        .read_text(datasets['train'], blocksize=10*MEGABYTES)
        .map(json.loads)
        .map(lambda x: {
            'published_timestamp': x.get('published', {}).get('$date'),
            'read_time': x.get('meta_tags', {}).get('twitter:data1').split(' ')[0]
        })
        .to_dataframe({
            'published_timestamp': 'datetime64[ns]',
            'read_time': int
        })
    )
    
    df = dag.compute()
    
dg.visualize([rprof, prof])
```
