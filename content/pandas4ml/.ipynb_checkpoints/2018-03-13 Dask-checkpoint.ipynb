````python
home = '/home/severo/.kaggle/datasets/mczielinski/bitcoin-historical-data/'
files = !ls {home}
datasets = {f.split('_')[0]:home+f for f in files if f.endswith('.csv')}
datasets
````

# Pandas

````python
(
    pd
    .read_csv(datasets['bitflyerJPY'])
    .assign(timestamp=lambda df: pd.to_datetime(df['Timestamp'], unit='s'))
    .assign(price=lambda df: df['Close'])
    .set_index('timestamp')
    ['price']
    .plot(figsize=(8,4), linewidth=1.5)
);
````

````python
sum(pd.read_csv(f).memory_usage().sum() for f in datasets.values())/MEGABYTES
````
